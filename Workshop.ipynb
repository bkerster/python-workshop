```python
def calc_xflips(x):
    #get direction of movement; 1 forward, -1 backward
    directions = []
    for i, num in enumerate(x[1:]):
        i += 1
        if x[i] > x[i-1]:
            directions.append(1)
        elif x[i] < x[i-1]:
            directions.append(-1)
        elif x[i] == x[i-1]:
            directions.append(0)
    
    # check if the direction changed
    num_flips = 0 
    current_dir = 0
    for dirr in directions:
        if current_dir == 1 and dirr == -1:
            num_flips += 1
        elif current_dir == -1 and dirr == 1:
            num_flips += 1
        if dirr != 0:
            current_dir = dirr
    
    return num_flips
```

```python
# lets test our function
xvals = [0, 1, 2,3, 4, 5, 6, 5, 4, 3, 2, 2, 2, 2, 1, 0, 1, 2, 3]
calc_xflips(xvals)
```
