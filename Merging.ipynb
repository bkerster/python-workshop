```python
import numpy as np
import pandas as pd
import os
```

# Place all 10 files into a single dataframe

## First lets get all the CSV files in the folder

```python
files = []
for f_name in os.listdir('.'):
    if f_name[-4:] == '.csv':
        files.append(f_name)
print( files )
```

## Hiccup #1: The files have different lengths. 

control_left	720 <br>
control_right	720<br>
focal_left	560<br>
focal_left_pm	80<br>
focal_right	392<br>
focal_right_pm	56<br>
nonfocal_left	616<br>
nonfocal_left_pm	80<br>
nonfocal_right	560<br>
nonfocal_right_pm	80<br>

### Let's put those lengths in a list. We'll need them shortly.

```python
table_lengths = [ 720, 720, 560, 80, 392, 56, 616, 80, 560, 80]
```

```python
all_dfs = [] # list store the dataframes in
for i, f_name in enumerate(files):
    temp_df = pd.read_csv(f_name, skiprows=1, nrows=table_lengths[i])
    # make sure to keep track of what condition this is
    temp_df['file_name'] = f_name[:-4]
    
    all_dfs.append(temp_df)
    
```

### Do a quick sanity check

```python
all_dfs[0].tail()
```

```python
all_dfs[1].head()
```

```python
df = pd.concat(all_dfs, ignore_index=True)
```

# Place the raw mouse data into arrays

### <p>We essentially want to perform one operation on each row<br></p> <p>X_1....X_101 -> np.array(x)<br></p> <p>Y_1....Y_101 -> np.array(y)</p>

```python
# lets write a function that performs this operation on a single row
def convert_ts_to_array(row):
    x = row['X_1':'X_101']
    y = row['Y_1':'Y_101']
    
    return pd.Series({'ts_x':np.array(x), 'ts_y':np.array(y)})
```

```python
# Let's test our function
convert_ts_to_array(df.ix[0])
```

### You can't add two columns to a dataframe using a single apply

```python
temp_ts = df.apply(convert_ts_to_array, axis=1)
```

```python
temp_ts.head()
```

### <p>We now have a dataframe containing our two time series columns</p> <p>This temp dataframe shares the same index as our main dataframe, so we'll merge them</p>

```python
df = pd.merge(df, temp_ts, left_index=True, right_index=True)
```
